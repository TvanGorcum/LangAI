```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'lai-data/nationality.csv'
df = pd.read_csv(file_path)
```

```python
df.info()
```

```python
nationality_counts = df['nationality'].value_counts()

plt.figure(figsize=(12, 8))
# Plot the bar chart
nationality_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Customize the plot

plt.title('Count of Each Nationality')
plt.xlabel('Nationality')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Group by 'nationality' and count the number of unique authors
nationality_author_counts = df.groupby('nationality')['auhtor_ID'].nunique()

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot the bar chart
nationality_author_counts.sort_values(ascending=False).plot(kind='bar', color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Number of Unique Authors from Each Country')
plt.xlabel('Country')
plt.ylabel('Number of Unique Authors')

# Show the plot
plt.show()
```
