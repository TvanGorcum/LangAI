```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#Predict according
df['predicted_nationality_germanic'] = np.random.choice(['Germanic', 'Non-Germanic'], len(df), p=[19743/59124, 1-(19743/59124)])
df['predicted_nationality_romance'] = np.random.choice(['Romance', 'Non-Romance'], len(df), p=[9333/59214, 1-(9333/59214)])

#labels for Germanic and Romance languages
germanic_labels = ['Germany', 'The Netherlands']
romance_labels = ['France', 'Italy', 'Spain', 'Portugal']

df['is_germanic'] = df['nationality'].isin(germanic_labels)
df['is_romance'] = df['nationality'].isin(romance_labels)

#Calculating the metrics for Germanic languages
accuracy_germanic = accuracy_score(df['is_germanic'], df['predicted_nationality_germanic'] == 'Germanic')
precision_germanic = precision_score(df['is_germanic'], df['predicted_nationality_germanic'] == 'Germanic')
recall_germanic = recall_score(df['is_germanic'], df['predicted_nationality_germanic'] == 'Germanic')
f1_germanic = f1_score(df['is_germanic'], df['predicted_nationality_germanic'] == 'Germanic')

#Calculating the metrics for Romance languages
accuracy_romance = accuracy_score(df['is_romance'], df['predicted_nationality_romance'] == 'Romance')
precision_romance = precision_score(df['is_romance'], df['predicted_nationality_romance'] == 'Romance')
recall_romance = recall_score(df['is_romance'], df['predicted_nationality_romance'] == 'Romance')
f1_romance = f1_score(df['is_romance'], df['predicted_nationality_romance'] == 'Romance')

print("Metrics for Germanic languages:")
print(f"Accuracy: {accuracy_germanic:.2f}")
print(f"Precision: {precision_germanic:.2f}")
print(f"Recall: {recall_germanic:.2f}")
print(f"F1 Score: {f1_germanic:.2f}")

print("\nMetrics for Romance languages:")
print(f"Accuracy: {accuracy_romance:.2f}")
print(f"Precision: {precision_romance:.2f}")
print(f"Recall: {recall_romance:.2f}")
print(f"F1 Score: {f1_romance:.2f}")
```
